```python
from __future__ import unicode_literals
scale = 5
fig,ax = subplots(figsize=(scale,scale))
a , b = randn(500),rand(500)
ax.scatter(x=a,y=b,c=linspace(0,1,len(a)))#,s=randn(500)*100, alpha=0.5)
ax.set_title('mi título')
ax.set_xlabel('var1')
ax.set_ylabel('var2')
ax.text(0.075, 1- 0.9*(0.1*4/3.),"text",fontsize=35,
                verticalalignment='center', ha="center",
                transform=ax.transAxes,
                bbox=dict(edgecolor='black',facecolor='w', boxstyle='round,pad=0.25',alpha=.35),
                color='black')
```

### Quiz : Hagan un panel con disitintas distribuciones de probabilidad y mezclen dos utilizando distintos colores y disitintas transparencias (alphas)

```python
scale = 5
fig,ax = subplots(figsize=(scale,scale))
x = linspace(0,10)
ax.plot(x, sin(x))
ax.plot(x, cos(x))
ax.plot(x, sqrt(x))
```

\begin{equation*}
\begin{split}
\vec{B_c}(\vec{r},t;\vec{r}_0)=\frac{I(t)\mu_0}{2\pi}\frac{(x-x_0)\hat{y}-(y-y_0)\hat{x}}{(x-x_0)^2+(y-y_0)^2}
\end{split}
\end{equation*}

```python
xPoints=linspace(-15,15,256)
yPoints=linspace(-15,15,256)
X, Y  = np.meshgrid(xPoints,yPoints,indexing='ij')
```

```python
Bx = zeros_like(X)
By = zeros_like(Y)
def lineCurrent(I=1,x0=20,y0=20):
    global Bx, By
    r2 = (X-x0)**2 + (Y-y0)**2
    Bx += -I*(Y-y0)/r2
    By += I*(X-x0)/r2
```

```python
Bx = zeros_like(X)
By = zeros_like(Y)
f=figure(1,figsize=(2,2),dpi=200)
sampling = 4
intense = 450
lineCurrent(intense,x0=26,y0=26)
lineCurrent(intense,x0=-26,y0=-26)
magB = sqrt(Bx*Bx+By*By)
magn = imshow(magB,origin='lower',cmap='inferno',extent=[xPoints[0],xPoints[-1],yPoints[0],yPoints[-1]])
cb  = colorbar(magn, ax=f.axes,orientation='vertical',fraction=0.045, pad=0.02,shrink=0.95)
```
